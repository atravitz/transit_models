```python
import transit_model as tm
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

```python
g, passengers = tm.initialize(n_passengers=100, node_capacity=1000)
```

```python
node_coords = {}
for n in g:
    node_coords[n] = g.nodes[n]['pos']
colors = [g.nodes[i]['population'] for i in g.nodes]
```

```python
plt.figure(dpi=120)
nx.draw(g, node_coords,
        alpha=0.5,
        node_size=20,
        node_color=[g.nodes[i]['nodecolor'] for i in g.nodes])
plt.axis('equal')
plt.show()
```
