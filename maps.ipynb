```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
l_stops = pd.read_csv("CTA_-_System_Information_-_List_of__L__Stops_-_Map.csv")
```

```python
l_stops
```

```python
coords = pd.DataFrame(l_stops.Location.apply(lambda x: x[1:-1].split(",")).tolist(), columns=['lat', 'long']).astype('float')
```

```python
fig = plt.figure(dpi=120)
plt.scatter(coords.long, coords.lat, alpha=0.3)
plt.axis('equal')
plt.grid()
plt.show()
```
