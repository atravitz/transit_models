```python
import transit_model as tm
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
import seaborn as sns
import networkx as nx
```

```python
g_CTA, passengers = tm.initialize(n_passengers=0, node_capacity=0)
```

```python
g_CTA1, passengers = tm.initialize(n_passengers=0, node_capacity=0)
g_CTA1 = tm.add_line('CTA_data/new_line_1.txt', g_CTA1, 'lightblue')
```

```python
node_coords = {}
for n in g_CTA:
    node_coords[n] = g_CTA.nodes[n]['pos']
colors = [g_CTA.nodes[i]['population'] for i in g_CTA.nodes]
```

```python
plt.figure(dpi=200)
nx.draw(g_CTA, node_coords,
        alpha=0.5,
        node_size=20,
        node_color=list(nx.betweenness_centrality(g_CTA).values()))
plt.axis('equal')
plt.show()

plt.figure(dpi=200)
nx.draw(g_CTA1, node_coords,
        alpha=0.5,
        node_size=20,
        node_color=list(nx.betweenness_centrality(g_CTA1).values()))
plt.axis('equal')
plt.show()
```
